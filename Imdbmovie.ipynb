### Removing irrelevant column

```python
df = df.drop(columns=['tagline','production_countries','overview','poster_path'])
df.head(1)
```

### Renaming columns

```python
df = df.rename(columns={
    'budget_musd':'budget',
    'revenue_musd':'revenue',
    'vote_average':'rating'
})
df.head(1)
```

```python
df.isnull().sum()
```

### Filling null values with average

```python
df = df.fillna({
    'budget':df['budget'].mean(),
    'revenue':df['revenue'].mean(),
    'rating':df['rating'].mean(),
    'runtime':df['runtime'].mean() 
}                           
)
df.isnull().sum()
```

### Creating new columns Profit & roi

```python
df['profit'] = df['revenue']-df['budget']
df['roi'] = df['revenue']/df['budget']
df.head(1)
```

### Best and worst movies

```python
def best_find(col,num):
    return fdf.sort_values(by=col,ascending=False).head(num)

```

```python
# best_find('revenue',10)
# best_find('rating',10)
best_find('popularity',10)
```

```python
highest_budget = best_find('budget',20)
highest_budget
```

```python
highest_revenue = best_find('revenue',20)
highest_revenue
```

```python
highest_vote_count = best_find('vote_count',20)
highest_vote_count
```

### concate all best movies by creating a function

```python
edf = pd.DataFrame()
cols = list(fdf.columns)
cols.remove('title')
for col in cols:
    tdf = best_find(col,20)
    edf = pd.concat([edf,tdf])
    
edf
```

```python
edf.loc[edf['title'].duplicated(keep=False)]
```

```python
edf['title'].value_counts().head(10).to_csv('Best_movies.csv')
```

### Find worst movies

```python
def worst_find(col,num):
    return fdf.sort_values(by=col,ascending=True).head(num)
```

```python
edf = pd.DataFrame()
cols = list(fdf.columns)
cols.remove('title')
for col in cols:
    tdf = worst_find(col,20)
    edf = pd.concat([edf,tdf])
    
edf
```

```python
edf['title'].value_counts().head(10).to_csv('Worst_movies.csv')
```
